```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv('HR_Employee_Attrition_Data.csv')
df = pd.DataFrame(data)
df
```

```python
agg_data = df.groupby('Education')['MonthlyIncome'].mean().reset_index()
x = list(agg_data['Education'])
y = list(agg_data['MonthlyIncome'])
```

```python
plt.plot(x, y, marker='o', color='cyan')
plt.show()
```
